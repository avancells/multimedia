```python
import math
def rice_codifier(value, k):
    if value < 0:
        coded_rice = "0"
    else:
        coded_rice = "1"
    value = abs(value)
    m = 2 ** k
    c = int(math.ceil(math.log(m,2)))
    quotient = int(math.floor(value / m))
    reminder = value % m 
    
    div = int(math.pow(2,c) - m)
    
    for i in range(quotient):
        coded_rice += "1"
    
    if (reminder < div):
        b = c - 1
        a = "{0:0" + str(b) + "b}"
        last_bits = a.format(reminder)
    else:
        b = c
        a = "{0:0" + str(b) + "b}"
        last_bits = a.format(reminder + div)

    coded_rice +=  "0" +str(last_bits)
    return coded_rice  
```

**(a) ¿Cuántos bits son necesarios para codificar todos los enteros entre -1023 y +1023 (en codificación binaria natural con bit de signo)?**

```python
def rice_stats(k=5, values=1023):
    total_size_no_l0 = 0 # Amount of bits necessary to codify without leading zeros
    count = 0
    rice_size = 0
    rice_bits = []

    for i in range(-values, values + 1, 1):
        total_size_no_l0 += (i.bit_length() + 1)
        coded = rice_codifier(i, k)
        rice_size += len(coded) 
        rice_bits.append((i, coded))

        count += 1

    total_size_with_l0 = (i.bit_length() + 1) * count
    min_size_l0 = (i.bit_length() + 1)
    return total_size_with_l0, total_size_no_l0, min_size_l0, rice_size, rice_bits
```

```python
total_size_with_l0, total_size_no_l0, min_size_l0, rice_size, rice_bits = rice_stats()
print("Size using binary + sign + leading zeros: ",total_size_with_l0, " bits")
print("Size using binary + sign + no leading zeros: ",total_size_no_l0, " bits")
print("Size using Rice with k=5(m=32): ", rice_size, " bits")

print("Min. size(per number) binary + sign + leading zeros: ", min_size_l0)
```
